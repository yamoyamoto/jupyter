```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-1, 5, 0.01) #

T = 0.8

#　ステップ関数
y_1 = np.tanh(x/T)
y_2 = np.tanh(x/T) ** 2

# グラフの設定
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.grid()
plt.show()
```

```python
from mpl_toolkits.axes_grid.axislines import SubplotZero

x = np.arange(-1.2, 1.2, 0.01) #

B = 2

#　ステップ関数
y_1 = np.tanh(2*x)
y_2 = x
y_3 = np.tanh(0.8*x)


# グラフの設定
fig = plt.figure()
ax= fig.add_subplot(1,1,1)

ax.set_xticks([])
ax.set_yticks([])

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_xlabel("m", loc="right")
ax.set_ylabel("y", rotation=0, loc="top")

ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

print(ax.axis)

ax.plot(x, y_1, label="y=tanh(2*m)")
ax.plot(x, y_3, label="y=tanh(0.8*m)")
ax.plot(x, y_2, label="y=m")


plt.legend(bbox_to_anchor=(0, 1), loc='upper left', borderaxespad=0, fontsize=10)

plt.show()
```
