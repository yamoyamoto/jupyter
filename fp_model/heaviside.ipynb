```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-1, 5, 0.01) #

T = 0.8

#　ステップ関数
y_1 = np.tanh(x/T)
y_2 = np.tanh(x/T) ** 2

# グラフの設定
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.grid()
plt.show()
```
